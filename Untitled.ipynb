```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Создание DataFrame для анализа
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print("Первые 5 строк данных:")
print(df.head())
print(f"\nРазмерность данных: {X.shape}")
print(f"Названия признаков: {feature_names}")
print(f"Названия классов: {target_names}")
```

```python
# Оценка сбалансированности
def assess_balance(y):
    unique, counts = np.unique(y, return_counts=True)
    balance_ratio = counts.min() / counts.max()
    
    print("Оценка сбалансированности:")
    print(f"Количество образцов по классам: {dict(zip(unique, counts))}")
    print(f"Соотношение баланса: {balance_ratio:.3f}")
    
    if balance_ratio > 0.8:
        print("Датасет сбалансирован ✓")
    elif balance_ratio > 0.5:
        print("Датасет умеренно сбалансирован")
    else:
        print("Датасет несбалансирован ⚠")
    
    return balance_ratio

# Оценка достаточности
def assess_sufficiency(X, y):
    n_samples = X.shape[0]
    n_features = X.shape[1]
    n_classes = len(np.unique(y))
    
    print(f"\nОценка достаточности:")
    print(f"Количество образцов: {n_samples}")
    print(f"Количество признаков: {n_features}")
    print(f"Количество классов: {n_classes}")
    
    # Эмпирическое правило: минимум 10 samples на класс на feature
    min_samples_needed = n_classes * n_features * 10
    
    if n_samples >= min_samples_needed:
        print(f"Датасет достаточен для обучения ✓ (нужно: {min_samples_needed}, есть: {n_samples})")
        return True
    else:
        print(f"⚠ Внимание: данных может быть недостаточно (нужно: {min_samples_needed}, есть: {n_samples})")
        return False

# Выполнение оценок
balance_score = assess_balance(y)
sufficiency_score = assess_sufficiency(X, y)

# Визуализация распределения классов
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.countplot(x='species', data=df)
plt.title('Распределение классов')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
df['species'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Процентное распределение классов')

plt.tight_layout()
plt.show()
```

```python
# Валидация и предобработка данных
def validate_and_preprocess_data(X, y):
    # Проверка на NaN значения
    if np.isnan(X).any() or np.isnan(y).any():
        raise ValueError("Обнаружены NaN значения в данных")
    
    # Проверка на бесконечные значения
    if not np.isfinite(X).all():
        raise ValueError("Обнаружены бесконечные значения в данных")
    
    # Масштабирование признаков
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    return X_scaled, y, scaler

# Защита от переполнения памяти
def check_memory_usage(X, max_mb=100):
    memory_usage = X.nbytes / (1024 * 1024)  # в MB
    if memory_usage > max_mb:
        raise MemoryError(f"Данные занимают {memory_usage:.2f} MB, превышен лимит {max_mb} MB")
    print(f"Использование памяти: {memory_usage:.2f} MB ✓")
    return True

# Выполнение проверок
try:
    X_processed, y_processed, scaler = validate_and_preprocess_data(X, y)
    memory_ok = check_memory_usage(X_processed)
    
    # Разделение на тренировочную и тестовую выборки
    X_train, X_test, y_train, y_test = train_test_split(
        X_processed, y_processed, test_size=0.2, random_state=42, stratify=y
    )
    
    print(f"Размер тренировочной выборки: {X_train.shape}")
    print(f"Размер тестовой выборки: {X_test.shape}")
    
except Exception as e:
    print(f"Ошибка при обработке данных: {e}")
```

```python
# Создание и обучение модели
def create_and_train_model(X_train, y_train):
    # Инициализация модели
    model = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        max_depth=3  # Ограничение для предотвращения переобучения
    )
    
    # Обучение модели
    model.fit(X_train, y_train)
    
    return model

# Обучение модели
model = create_and_train_model(X_train, y_train)

# Кросс-валидация для оценки стабильности модели
cv_scores = cross_val_score(model, X_processed, y_processed, cv=5)
print(f"Результаты кросс-валидации (5-fold): {cv_scores}")
print(f"Средняя точность кросс-валидации: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
```

```python
def validate_input_data(X, expected_features=4):
    # Проверка типа данных
    if not isinstance(X, (np.ndarray, list)):
        raise TypeError("Входные данные должны быть numpy array или list")
    
    # Преобразование в numpy array если необходимо
    X_array = np.array(X)
    
    # Проверка формы данных
    if len(X_array.shape) != 2:
        raise ValueError("Данные должны быть двумерным массивом")
    
    if X_array.shape[1] != expected_features:
        raise ValueError(f"Ожидается {expected_features} признаков, получено {X_array.shape[1]}")
    
    # Проверка на NaN
    if np.isnan(X_array).any():
        raise ValueError("Обнаружены NaN значения во входных данных")
    
    # Проверка на бесконечные значения
    if not np.isfinite(X_array).all():
        raise ValueError("Обнаружены бесконечные значения во входных данных")
    
    return X_array
```

```python
# Предсказания и оценка качества
y_pred = model.predict(X_test)

print("=" * 50)
print("ОЦЕНКА КАЧЕСТВА МОДЕЛИ")
print("=" * 50)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.3f}")

# Подробный отчет
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))

# Матрица ошибок
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=target_names, yticklabels=target_names)
plt.title('Матрица ошибок')
plt.ylabel('Истинные значения')
plt.xlabel('Предсказанные значения')
plt.show()

# Важность признаков
feature_importance = model.feature_importances_
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance, y=feature_names)
plt.title('Важность признаков')
plt.xlabel('Важность')
plt.tight_layout()
plt.show()
```

```python
# Визуализация предсказаний
def visualize_predictions(X_test, y_test, y_pred, feature_names):
    # Сокращение размерности для визуализации (используем первые 2 признака)
    plt.figure(figsize=(15, 5))
    
    # Истинные значения
    plt.subplot(1, 3, 1)
    scatter = plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='viridis', alpha=0.7)
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title('Истинные классы')
    plt.colorbar(scatter, ticks=[0, 1, 2]).set_ticklabels(target_names)
    
    # Предсказанные значения
    plt.subplot(1, 3, 2)
    scatter = plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='viridis', alpha=0.7)
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title('Предсказанные классы')
    plt.colorbar(scatter, ticks=[0, 1, 2]).set_ticklabels(target_names)
    
    # Ошибки классификации
    plt.subplot(1, 3, 3)
    errors = y_test != y_pred
    correct = y_test == y_pred
    plt.scatter(X_test[correct, 0], X_test[correct, 1], c='green', alpha=0.5, label='Правильно')
    plt.scatter(X_test[errors, 0], X_test[errors, 1], c='red', alpha=0.8, label='Ошибки')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title('Ошибки классификации')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Визуализация
visualize_predictions(X_test, y_test, y_pred, feature_names)

# Дополнительная статистика
print("\nДОПОЛНИТЕЛЬНАЯ СТАТИСТИКА:")
print(f"Количество правильных предсказаний: {np.sum(y_test == y_pred)}/{len(y_test)}")
print(f"Количество ошибок: {np.sum(y_test != y_pred)}")
```

```python
import joblib
import os
from pathlib import Path

def save_model_simple(model, scaler, model_name="iris_model"):
    """Простое сохранение модели с безопасностью"""
    # Создаем директорию
    model_dir = Path("./models")
    model_dir.mkdir(exist_ok=True)
    
    # Сохраняем модель и scaler
    joblib.dump(model, model_dir / f"{model_name}.joblib")
    joblib.dump(scaler, model_dir / f"{model_name}_scaler.joblib")
    
    # Защищаем права доступа (только для владельца)
    os.chmod(model_dir / f"{model_name}.joblib", 0o600)
    os.chmod(model_dir / f"{model_name}_scaler.joblib", 0o600)
    
    print(f"✓ Модель сохранена: ./models/{model_name}.joblib")
    return True

# Сохраняем модель
save_model_simple(model, scaler, "iris_classifier")
```

```python

def load_model_simple(model_name="iris_model"):
    """Простая загрузка модели с проверками"""
    model_path = f"./models/{model_name}.joblib"
    scaler_path = f"./models/{model_name}_scaler.joblib"
    
    # Проверяем существование файлов
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"Модель {model_path} не найдена")
    
    # Загружаем модель и scaler
    model = joblib.load(model_path)
    scaler = joblib.load(scaler_path)
    
    print("✓ Модель успешно загружена")
    return model, scaler

# Загружаем модель
loaded_model, loaded_scaler = load_model_simple("iris_classifier")
```
